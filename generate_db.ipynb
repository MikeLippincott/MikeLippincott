```python
import sqlite3
from pathlib import Path

# Configure database location
db_path = Path("cv_raw_data.db")

# Create database + table for raw CV content
conn = sqlite3.connect(db_path)
cur = conn.cursor()

cur.execute(
    """
CREATE TABLE IF NOT EXISTS raw_cv (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    candidate_name TEXT,
    source_file TEXT NOT NULL,
    raw_text TEXT NOT NULL,
    imported_at TEXT DEFAULT (datetime('now')),
    parse_status TEXT DEFAULT 'pending',
    notes TEXT
)
"""
)

# Optional index for faster status filtering
cur.execute(
    """
CREATE INDEX IF NOT EXISTS idx_raw_cv_parse_status
ON raw_cv(parse_status)
"""
)

conn.commit()
conn.close()

print(f"SQLite DB ready: {db_path.resolve()}")
```
